```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

data = pd.read_csv("https://raw.githubusercontent.com/shreythapar/ML_Dataset/main/Disease%20Prediction.csv?token=APJ435FPHHD6BXZTUI7WEQTA6WMV4", 
                  usecols=["itching", "skin_rash", "prognosis", "cough", "nodal_skin_eruptions", "joint_pain"])
```

```python
from sklearn import preprocessing 
#before normalisation
print(data['itching'].plot(kind = 'bar'))
```

```python
df_min_max_scaled = data.copy()
# apply normalization techniques to column 'ExitRates'
column = 'itching'
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min()) 
# view normalized data
print(df_min_max_scaled['itching'].plot(kind='bar'))
```

```python
le = preprocessing.LabelEncoder()
data['skin_rash'] = le.fit_transform(data['skin_rash'])
# as you can see the data has been encoded 
data['skin_rash'].head()
```

```python
print(data.skew(axis = 0, skipna = True))

data.joint_pain.hist()
plt.title('joint_pain')
```

```python
#function to return plots for the feature
#importing necessary libraries
import scipy.stats as stats
import pylab
import seaborn as sns
def normality(data,feature):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(data[feature])
    plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()

#performing logarithmic transformation on the feature
data['joint_pain']=np.log(data['joint_pain'])
#plotting to check the transformation
normality(data,'joint_pain')
```

```python
# using feature engineering wherever necessary 
# normalising all data with large values and are of float datatype

#performing logarithmic transformation on the feature
data['itching']=np.log(data['itching'])
#plotting to check the transformation
normality(data,'itching')
```

```python
#performing logarithmic transformation on the feature
data['skin_rash']=np.log(data['skin_rash'])
#plotting to check the transformation
normality(data,'skin_rash')
```
