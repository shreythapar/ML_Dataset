```python
import pandas as pd
from matplotlib import pyplot as plt

df2 = pd.read_csv("https://raw.githubusercontent.com/shreythapar/ML_Dataset/main/Disease%20Prediction.csv?token=APJ435FPHHD6BXZTUI7WEQTA6WMV4", 
                  usecols=["itching", "skin_rash", "prognosis", "cough", "nodal_skin_eruptions", "joint_pain"])

plt.figure(figsize = (2, 20))
plt.scatter(df2['itching'], df2['prognosis'])
plt.xlabel('Itching')
plt.ylabel('Skin Rash')
plt.title("Itching v/s Skin Rash")
plt.show()
```

```python
plt.figure(figsize = (4, 4))
plt.bar(df2['itching'], df2['skin_rash'])
plt.xlabel('Itching')
plt.ylabel('Skin Rash')
plt.title("Itching v/s Skin Rash")
plt.show()
```

```python
# plt.figure(figsize = (4, 4))
plt.hist(df2['itching'], bins = 15)
plt.title("Itching v/s Skin Rash")
plt.xlabel('Itching')
plt.ylabel('Skin Rash')
plt.show()
```

```python
import seaborn as sns
sns.countplot(y = df2['prognosis'], data=df2)
sns.set(rc={'figure.figsize':(61.7,8.27)})
```
